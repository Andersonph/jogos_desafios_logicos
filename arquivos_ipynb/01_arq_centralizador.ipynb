```python
import pygame
import sys
import random

pygame.init()

# =========================
# CONFIGURAÇÕES PRINCIPAIS
# =========================

LARGURA = 1000
ALTURA = 650

tela = pygame.display.set_mode((LARGURA, ALTURA))
pygame.display.set_caption("Desafios Lógicos")

fonte = pygame.font.SysFont(None, 32)
fonte_titulo = pygame.font.SysFont(None, 52)
fonte_quebra = pygame.font.SysFont(None, 60)

estado = "menu_principal"


# =========================
# FUNÇÃO GERAL
# =========================

def desenhar_texto(texto, x, y, fonte_usada=fonte, cor=(0, 0, 0)):
    img = fonte_usada.render(texto, True, cor)
    tela.blit(img, (x, y))


# =========================
# DESAFIO 1 - TRAVESSIA
# =========================

esquerda = ["Raposa", "Galinha", "Alface", "Fazendeiro"]
direita = []

lado_fazendeiro = "esquerda"

movimentos = 0

mensagem_vitoria = ""
mensagem_derrota = ""


def reiniciar_travessia():
    global esquerda
    global direita
    global lado_fazendeiro
    global movimentos
    global mensagem_vitoria
    global mensagem_derrota
    global estado

    esquerda = ["Raposa", "Galinha", "Alface", "Fazendeiro"]
    direita = []

    lado_fazendeiro = "esquerda"

    movimentos = 0

    mensagem_vitoria = ""
    mensagem_derrota = ""

    estado = "travessia_menu"


def verificar_estado():
    global estado
    global mensagem_vitoria
    global mensagem_derrota

    if (
        "Raposa" in esquerda
        and "Galinha" in esquerda
        and "Fazendeiro" not in esquerda
    ):
        estado = "derrota"
        mensagem_derrota = "A raposa comeu a galinha!"
        return

    if (
        "Raposa" in direita
        and "Galinha" in direita
        and "Fazendeiro" not in direita
    ):
        estado = "derrota"
        mensagem_derrota = "A raposa comeu a galinha!"
        return

    if (
        "Galinha" in esquerda
        and "Alface" in esquerda
        and "Fazendeiro" not in esquerda
    ):
        estado = "derrota"
        mensagem_derrota = "A galinha comeu a alface!"
        return

    if (
        "Galinha" in direita
        and "Alface" in direita
        and "Fazendeiro" not in direita
    ):
        estado = "derrota"
        mensagem_derrota = "A galinha comeu a alface!"
        return

    if len(direita) == 4:
        estado = "vitoria"
        mensagem_vitoria = (
            f"Parabéns! Você venceu em {movimentos} movimentos."
        )


def atravessar(item=None):
    global lado_fazendeiro
    global movimentos

    if estado != "travessia_jogando":
        return

    if lado_fazendeiro == "esquerda":

        if item:
            if item not in esquerda:
                return

            esquerda.remove(item)
            direita.append(item)

        esquerda.remove("Fazendeiro")
        direita.append("Fazendeiro")

        lado_fazendeiro = "direita"

    else:

        if item:
            if item not in direita:
                return

            direita.remove(item)
            esquerda.append(item)

        direita.remove("Fazendeiro")
        esquerda.append("Fazendeiro")

        lado_fazendeiro = "esquerda"

    movimentos += 1

    verificar_estado()
    

# =========================
# DESAFIO 2 - QUEBRA-CABEÇA
# =========================

TAMANHO_GRADE = 4
TAMANHO_PECA = 90

INICIO_X = 320
INICIO_Y = 120

grade_quebra = []
movimentos_quebra = 0


def embaralhar_grade():
    grade = [
        [
            (i * TAMANHO_GRADE + j + 1) % (TAMANHO_GRADE * TAMANHO_GRADE)
            for j in range(TAMANHO_GRADE)
        ]
        for i in range(TAMANHO_GRADE)
    ]

    movimentos_possiveis = [(0, 1), (0, -1), (1, 0), (-1, 0)]

    vazio_l = TAMANHO_GRADE - 1
    vazio_c = TAMANHO_GRADE - 1

    for _ in range(100):
        random.shuffle(movimentos_possiveis)

        for dl, dc in movimentos_possiveis:
            novo_l = vazio_l + dl
            novo_c = vazio_c + dc

            if 0 <= novo_l < TAMANHO_GRADE and 0 <= novo_c < TAMANHO_GRADE:
                grade[vazio_l][vazio_c] = grade[novo_l][novo_c]
                grade[novo_l][novo_c] = 0

                vazio_l = novo_l
                vazio_c = novo_c
                break

    return grade


def reiniciar_quebra_cabeca():
    global grade_quebra
    global movimentos_quebra
    global estado

    grade_quebra = embaralhar_grade()
    movimentos_quebra = 0

    estado = "quebra_menu"


def desenhar_grade_quebra():
    pygame.draw.rect(
        tela,
        (0, 0, 0),
        (
            INICIO_X - 5,
            INICIO_Y - 5,
            TAMANHO_PECA * 4 + 10,
            TAMANHO_PECA * 4 + 10,
        ),
    )

    for linha in range(TAMANHO_GRADE):
        for coluna in range(TAMANHO_GRADE):
            valor = grade_quebra[linha][coluna]

            x = INICIO_X + coluna * TAMANHO_PECA
            y = INICIO_Y + linha * TAMANHO_PECA

            if valor != 0:
                pygame.draw.rect(
                    tela,
                    (50, 150, 250),
                    (x + 2, y + 2, TAMANHO_PECA - 4, TAMANHO_PECA - 4),
                )

                texto = fonte_quebra.render(str(valor), True, (255, 255, 255))
                rect_texto = texto.get_rect(
                    center=(
                        x + TAMANHO_PECA // 2,
                        y + TAMANHO_PECA // 2,
                    )
                )

                tela.blit(texto, rect_texto)


def obter_posicao_vazio():
    for l in range(TAMANHO_GRADE):
        for c in range(TAMANHO_GRADE):
            if grade_quebra[l][c] == 0:
                return l, c


def mover_peca_quebra(clique_l, clique_c):
    vazio_l, vazio_c = obter_posicao_vazio()

    if (
        abs(clique_l - vazio_l) == 1
        and clique_c == vazio_c
    ) or (
        abs(clique_c - vazio_c) == 1
        and clique_l == vazio_l
    ):
        grade_quebra[vazio_l][vazio_c], grade_quebra[clique_l][clique_c] = (
            grade_quebra[clique_l][clique_c],
            grade_quebra[vazio_l][vazio_c],
        )

        return True

    return False


def verificar_vitoria_quebra():
    valor_esperado = 1

    for l in range(TAMANHO_GRADE):
        for c in range(TAMANHO_GRADE):
            if l == TAMANHO_GRADE - 1 and c == TAMANHO_GRADE - 1:
                return grade_quebra[l][c] == 0

            if grade_quebra[l][c] != valor_esperado:
                return False

            valor_esperado += 1

    return True

# =========================
# DESAFIO 3 - ENIGMAS
# =========================

enigma_pool = [
    ("O que nasce grande e morre pequeno?", "O lapis", "O ser humano", "A arvore"),
    ("Tenho chaves, mas nao abro portas. Tenho espaco, mas nao tenho salas?", "O teclado", "O hotel", "O chaveiro"),
    ("O que corre mas nao tem pernas, e tem leito mas nao dorme?", "O rio", "O vento", "O trem"),
    ("Faco duas pessoas a partir de uma. O que eu sou?", "O espelho", "A sombra", "A foto"),
    ("Se voce me soltar eu quebro, mas se sorrir para mim eu sorrio de volta.", "O espelho", "O vidro", "A agua"),
    ("O que esta sempre a caminho, mas nunca chega de fato?", "O amanha", "O carteiro", "O futuro distante"),
    ("Quanto mais voce tira de mim, maior eu fico. O que eu sou?", "Um buraco", "A areia", "A idade"),
    ("O que tem muitas palavras, mas nunca fala nada?", "O livro", "O dicionario", "A mente"),
    ("Estou sempre no passado, criado no presente, mudo o futuro.", "A escolha", "O vento", "O pensamento"),
    ("Tenho cidades, mas nao casas. Tenho montanhas, mas nao arvores.", "O mapa", "O espaco", "O deserto"),
]

perguntas_enigma = []
pergunta_atual_enigma = None
respostas_enigma = []

indice_enigma = 0
pontuacao_enigma = 0

resposta_escolhida = None
mostrar_resultado = False

TOTAL_ENIGMAS = 5

botoes_respostas_enigma = []

botao_proxima_enigma = pygame.Rect(390, 520, 220, 50)


def reiniciar_enigma():
    global perguntas_enigma
    global indice_enigma
    global pontuacao_enigma
    global resposta_escolhida
    global mostrar_resultado
    global estado

    perguntas_enigma = list(enigma_pool)
    random.shuffle(perguntas_enigma)

    indice_enigma = 0
    pontuacao_enigma = 0

    resposta_escolhida = None
    mostrar_resultado = False

    carregar_pergunta_enigma()

    estado = "enigma_menu"


def carregar_pergunta_enigma():
    global pergunta_atual_enigma
    global respostas_enigma
    global botoes_respostas_enigma
    global resposta_escolhida
    global mostrar_resultado

    resposta_escolhida = None
    mostrar_resultado = False

    pergunta, correta, errada1, errada2 = perguntas_enigma[indice_enigma]

    respostas_enigma = [correta, errada1, errada2]
    random.shuffle(respostas_enigma)

    indice_correto = respostas_enigma.index(correta)

    pergunta_atual_enigma = (pergunta, respostas_enigma, indice_correto)

    botoes_respostas_enigma = [
        pygame.Rect(250, 260, 500, 55),
        pygame.Rect(250, 335, 500, 55),
        pygame.Rect(250, 410, 500, 55),
    ]


def verificar_resposta_enigma(indice):
    global resposta_escolhida
    global mostrar_resultado
    global pontuacao_enigma

    if mostrar_resultado:
        return

    resposta_escolhida = indice
    mostrar_resultado = True

    if indice == pergunta_atual_enigma[2]:
        pontuacao_enigma += 10


def proxima_pergunta_enigma():
    global indice_enigma
    global estado

    indice_enigma += 1

    if indice_enigma >= TOTAL_ENIGMAS:
        estado = "enigma_fim"
    else:
        carregar_pergunta_enigma()

# =========================
# BOTÕES
# =========================

botoes_menu = {
    "Desafio 1 - Travessia": pygame.Rect(330, 180, 340, 50),
    "Desafio 2 - Quebra-cabeça": pygame.Rect(330, 250, 340, 50),
    "Desafio 3 - Enigmas": pygame.Rect(330, 320, 340, 50),
    "Desafio 4": pygame.Rect(330, 390, 340, 50),
    "Sair": pygame.Rect(330, 480, 340, 50),
}

botoes_travessia = {
    "Raposa": pygame.Rect(380, 140, 220, 45),
    "Galinha": pygame.Rect(380, 200, 220, 45),
    "Alface": pygame.Rect(380, 260, 220, 45),
    "Sozinho": pygame.Rect(380, 320, 220, 45),
}

# =========================
# LOOP PRINCIPAL
# =========================

while True:
    tela.fill((173, 216, 230))

    # =========================
    # TELAS
    # =========================

    if estado == "menu_principal":
        desenhar_texto("DESAFIOS LÓGICOS", 320, 60, fonte_titulo)
        desenhar_texto("Escolha um desafio para jogar:", 330, 130)

        for nome, rect in botoes_menu.items():
            pygame.draw.rect(tela, (220, 220, 220), rect)
            desenhar_texto(nome, rect.x + 25, rect.y + 13)

    elif estado == "travessia_menu":
        desenhar_texto("TRAVESSIAS IMPOSSÍVEIS", 250, 40, fonte_titulo)
        desenhar_texto("O fazendeiro precisa atravessar o rio.", 170, 140)
        desenhar_texto("Regras:", 170, 190)
        desenhar_texto("A raposa não pode ficar sozinha com a galinha.", 170, 250)
        desenhar_texto("A galinha não pode ficar sozinha com a alface.", 170, 300)
        desenhar_texto("O barco leva o fazendeiro e mais um passageiro.", 170, 350)
        desenhar_texto("Pressione ENTER para iniciar.", 170, 470)
        desenhar_texto("Pressione ESC para voltar ao menu principal.", 170, 520)

    elif estado == "travessia_jogando":
        pygame.draw.line(tela, (0, 0, 255), (500, 0), (500, ALTURA), 5)

        desenhar_texto("Margem Esquerda", 120, 40)
        desenhar_texto("Margem Direita", 700, 40)

        y = 120
        for item in esquerda:
            desenhar_texto(item, 120, y)
            y += 45

        y = 120
        for item in direita:
            desenhar_texto(item, 700, y)
            y += 45

        for nome, rect in botoes_travessia.items():
            pygame.draw.rect(tela, (220, 220, 220), rect)
            desenhar_texto(nome, rect.x + 20, rect.y + 10)

        desenhar_texto(f"Movimentos: {movimentos}", 400, 420)
        desenhar_texto(f"Barco na: {lado_fazendeiro}", 380, 470)
        desenhar_texto("ESC - Voltar ao menu", 380, 530)

    elif estado == "vitoria":
        desenhar_texto("PARABÉNS!", 360, 180, fonte_titulo)
        desenhar_texto(mensagem_vitoria, 220, 280)
        desenhar_texto("Pressione R para jogar novamente", 220, 380)
        desenhar_texto("Pressione ESC para voltar ao menu", 220, 430)

    elif estado == "derrota":
        desenhar_texto("VOCÊ PERDEU!", 330, 180, fonte_titulo)
        desenhar_texto(mensagem_derrota, 280, 280)
        desenhar_texto("Pressione R para tentar novamente", 220, 380)
        desenhar_texto("Pressione ESC para voltar ao menu", 220, 430)

    elif estado == "quebra_menu":
        desenhar_texto("QUEBRA-CABEÇA NUMÉRICO", 240, 60, fonte_titulo)
        desenhar_texto("Organize os números de 1 a 15.", 260, 170)
        desenhar_texto("Clique em uma peça ao lado do espaço vazio para mover.", 180, 230)
        desenhar_texto("Pressione ENTER para iniciar.", 300, 350)
        desenhar_texto("Pressione ESC para voltar ao menu principal.", 230, 410)

    elif estado == "quebra_jogando":
        desenhar_texto("QUEBRA-CABEÇA NUMÉRICO", 260, 40, fonte_titulo)
        desenhar_grade_quebra()
        desenhar_texto(f"Movimentos: {movimentos_quebra}", 390, 510)
        desenhar_texto("ESC - Voltar ao menu", 370, 560)

    elif estado == "quebra_vitoria":
        desenhar_texto("VITÓRIA!", 380, 180, fonte_titulo)
        desenhar_texto(f"Você resolveu em {movimentos_quebra} movimentos.", 280, 280)
        desenhar_texto("Pressione R para jogar novamente", 280, 380)
        desenhar_texto("Pressione ESC para voltar ao menu", 280, 430)

    elif estado == "enigma_menu":
        desenhar_texto("A PROFECIA DOS ENIGMAS", 250, 60, fonte_titulo)
        desenhar_texto("Responda aos enigmas e acumule pontos.", 250, 180)
        desenhar_texto("Cada acerto vale 10 pontos.", 250, 230)
        desenhar_texto("Pressione ENTER para iniciar.", 250, 350)
        desenhar_texto("Pressione ESC para voltar ao menu principal.", 250, 410)

    elif estado == "enigma_jogando":
        desenhar_texto("DESAFIO DOS ENIGMAS", 300, 40, fonte_titulo)
        desenhar_texto(f"Pergunta {indice_enigma + 1}/{TOTAL_ENIGMAS}", 400, 100)
        desenhar_texto(f"Pontuação: {pontuacao_enigma}", 400, 135)

        pergunta = pergunta_atual_enigma[0]
        desenhar_texto(pergunta, 150, 190)

        for i, rect in enumerate(botoes_respostas_enigma):
            cor = (220, 220, 220)

            if mostrar_resultado:
                if i == pergunta_atual_enigma[2]:
                    cor = (80, 200, 120)
                elif i == resposta_escolhida:
                    cor = (220, 80, 80)

            pygame.draw.rect(tela, cor, rect)
            desenhar_texto(respostas_enigma[i], rect.x + 20, rect.y + 15)

        if mostrar_resultado:
            pygame.draw.rect(tela, (220, 220, 220), botao_proxima_enigma)
            desenhar_texto(
                "Próxima",
                botao_proxima_enigma.x + 60,
                botao_proxima_enigma.y + 15,
            )

    elif estado == "enigma_fim":
        desenhar_texto("FIM DO DESAFIO!", 320, 170, fonte_titulo)
        desenhar_texto(
            f"Sua pontuação final foi: {pontuacao_enigma} pontos.",
            270,
            280,
        )
        desenhar_texto("Pressione R para jogar novamente", 270, 380)
        desenhar_texto("Pressione ESC para voltar ao menu", 270, 430)

    pygame.display.flip()

    # =========================
    # EVENTOS
    # =========================

    for evento in pygame.event.get():

        if evento.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

        if evento.type == pygame.KEYDOWN:

            if evento.key == pygame.K_ESCAPE:
                estado = "menu_principal"

            if estado == "travessia_menu" and evento.key == pygame.K_RETURN:
                estado = "travessia_jogando"

            elif estado in ["vitoria", "derrota"] and evento.key == pygame.K_r:
                reiniciar_travessia()

            elif estado == "quebra_menu" and evento.key == pygame.K_RETURN:
                estado = "quebra_jogando"

            elif estado == "quebra_vitoria" and evento.key == pygame.K_r:
                reiniciar_quebra_cabeca()

            elif estado == "enigma_menu" and evento.key == pygame.K_RETURN:
                estado = "enigma_jogando"

            elif estado == "enigma_fim" and evento.key == pygame.K_r:
                reiniciar_enigma()

        if evento.type == pygame.MOUSEBUTTONDOWN:

            if estado == "menu_principal":

                if botoes_menu["Desafio 1 - Travessia"].collidepoint(evento.pos):
                    reiniciar_travessia()

                elif botoes_menu["Desafio 2 - Quebra-cabeça"].collidepoint(evento.pos):
                    reiniciar_quebra_cabeca()

                elif botoes_menu["Desafio 3 - Enigmas"].collidepoint(evento.pos):
                    reiniciar_enigma()

                elif botoes_menu["Desafio 4"].collidepoint(evento.pos):
                    print("Desafio 4 ainda será criado.")

                elif botoes_menu["Sair"].collidepoint(evento.pos):
                    pygame.quit()
                    sys.exit()

            elif estado == "travessia_jogando":

                if botoes_travessia["Raposa"].collidepoint(evento.pos):
                    atravessar("Raposa")

                elif botoes_travessia["Galinha"].collidepoint(evento.pos):
                    atravessar("Galinha")

                elif botoes_travessia["Alface"].collidepoint(evento.pos):
                    atravessar("Alface")

                elif botoes_travessia["Sozinho"].collidepoint(evento.pos):
                    atravessar()

            elif estado == "quebra_jogando":

                x, y = evento.pos

                if (
                    INICIO_X <= x < INICIO_X + TAMANHO_PECA * 4
                    and INICIO_Y <= y < INICIO_Y + TAMANHO_PECA * 4
                ):
                    coluna = (x - INICIO_X) // TAMANHO_PECA
                    linha = (y - INICIO_Y) // TAMANHO_PECA

                    if mover_peca_quebra(linha, coluna):
                        movimentos_quebra += 1

                        if verificar_vitoria_quebra():
                            estado = "quebra_vitoria"

            elif estado == "enigma_jogando":

                for i, rect in enumerate(botoes_respostas_enigma):
                    if rect.collidepoint(evento.pos):
                        verificar_resposta_enigma(i)

                if (
                    mostrar_resultado
                    and botao_proxima_enigma.collidepoint(evento.pos)
                ):
                    proxima_pergunta_enigma()




```
