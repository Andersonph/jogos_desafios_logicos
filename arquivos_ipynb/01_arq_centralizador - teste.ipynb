```python
import pygame
import sys
import random

pygame.init()
pygame.mixer.init()

# =========================
# CONFIGURAÇÕES PRINCIPAIS
# =========================

LARGURA = 1000
ALTURA = 650

tela = pygame.display.set_mode((LARGURA, ALTURA))
pygame.display.set_caption("Desafios Lógicos")

fundo_menu = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_menu.png"
)

fundo_menu = pygame.transform.scale(
    fundo_menu,
    (LARGURA, ALTURA)
)

fundo_travessia_menu = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_travessia_menu.png"
)

fundo_travessia_menu = pygame.transform.scale(
    fundo_travessia_menu,
    (LARGURA, ALTURA)
)

fundo_travessia_jogando = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_travessia_jogando.png"
)

fundo_travessia_jogando = pygame.transform.scale(
    fundo_travessia_jogando,
    (LARGURA, ALTURA)
)

fundo_vitoria_travessia = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_vitoria_travessia.png"
)

fundo_vitoria_travessia = pygame.transform.scale(
    fundo_vitoria_travessia,
    (LARGURA, ALTURA)
)

fundo_derrota_travessia = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_derrota_travessia.png"
)

fundo_derrota_travessia = pygame.transform.scale(
    fundo_derrota_travessia,
    (LARGURA, ALTURA)
)

fundo_quebra_menu = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_quebra_menu.png"
)

fundo_quebra_menu = pygame.transform.scale(
    fundo_quebra_menu,
    (LARGURA, ALTURA)
)

fundo_quebra_jogando = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_quebra_jogando.png"
)

fundo_quebra_jogando = pygame.transform.scale(
    fundo_quebra_jogando,
    (LARGURA, ALTURA)
)

fundo_quebra_vitoria = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_quebra_vitoria.png"
)

fundo_quebra_vitoria = pygame.transform.scale(
    fundo_quebra_vitoria,
    (LARGURA, ALTURA)
)

fundo_quebra_derrota = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_quebra_derrota.png"
)

fundo_quebra_derrota = pygame.transform.scale(
    fundo_quebra_derrota,
    (LARGURA, ALTURA)
)

fundo_enigma_menu = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_enigma_menu.png"
)

fundo_enigma_menu = pygame.transform.scale(
    fundo_enigma_menu,
    (LARGURA, ALTURA)
)

fundo_enigma_jogando = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_enigma_jogando.png"
)

fundo_enigma_jogando = pygame.transform.scale(
    fundo_enigma_jogando,
    (LARGURA, ALTURA)
)

fundo_enigma_fim = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_enigma_fim.png"
)

fundo_enigma_fim = pygame.transform.scale(
    fundo_enigma_fim,
    (LARGURA, ALTURA)
)

fundo_space_menu = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_space_menu.png"
)

fundo_space_menu = pygame.transform.scale(
    fundo_space_menu,
    (LARGURA, ALTURA)
)

fundo_space_jogando = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_space_jogando.png"
)

fundo_space_jogando = pygame.transform.scale(
    fundo_space_jogando,
    (LARGURA, ALTURA)
)

asteroide_img = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\asteroide.png"
)

asteroide_img = pygame.transform.scale(
    asteroide_img,
    (90, 90)
)

fundo_space_derrota = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_space_derrota.png"
)

fundo_space_derrota = pygame.transform.scale(
    fundo_space_derrota,
    (LARGURA, ALTURA)
)


fundo_space_vitoria = pygame.image.load(
    r"C:\Git\jogos_desafios\imagens\fundo_space_vitoria.png"
)

fundo_space_vitoria = pygame.transform.scale(
    fundo_space_vitoria,
    (LARGURA, ALTURA)
)

# =========================
# SONS
# =========================

som_clique = pygame.mixer.Sound(
    r"C:\Git\jogos_desafios\sons\clique.mp3"
)

som_clique2 = pygame.mixer.Sound(
    r"C:\Git\jogos_desafios\sons\clique2.mp3"
)

som_vitoria = pygame.mixer.Sound(
    r"C:\Git\jogos_desafios\sons\vitoria.mp3"
)

som_derrota = pygame.mixer.Sound(
    r"C:\Git\jogos_desafios\sons\derrota.mp3"
)

# VOLUME DOS SONS
som_clique.set_volume(1)
som_clique2.set_volume(0.8)
som_vitoria.set_volume(0.50)
som_derrota.set_volume(0.50)

fonte = pygame.font.SysFont(None, 32)
fonte_titulo = pygame.font.SysFont(None, 52)
fonte_quebra = pygame.font.SysFont(None, 60)

estado = "menu_principal"


# =========================
# MÚSICAS
# =========================

musica_menu = r"C:\Git\jogos_desafios\sons\menu.mp3"
musica_travessia = r"C:\Git\jogos_desafios\sons\travessia.mp3"
musica_quebra = r"C:\Git\jogos_desafios\sons\quebra_cabeca.mp3"
musica_enigmas = r"C:\Git\jogos_desafios\sons\enigmas.mp3"
musica_space = r"C:\Git\jogos_desafios\sons\space_math.mp3"



# =========================
# FUNÇÃO GERAL
# =========================

musica_atual = None

def tocar_musica(caminho):
    global musica_atual

    if musica_atual != caminho:
        pygame.mixer.music.load(caminho)
        pygame.mixer.music.set_volume(0.30)
        pygame.mixer.music.play(-1)
        musica_atual = caminho

def desenhar_texto(texto, x, y, fonte_usada=fonte, cor=(0, 0, 0)):
    img = fonte_usada.render(texto, True, cor)
    tela.blit(img, (x, y))

def desenhar_texto_centralizado(texto, y, fonte_usada=fonte, cor=(0, 0, 0)):
    img = fonte_usada.render(texto, True, cor)

    rect = img.get_rect()
    rect.centerx = LARGURA // 2
    rect.y = y

    tela.blit(img, rect)

def desenhar_texto_multilinha(texto, y, largura_max=700,
                              fonte_usada=fonte, cor=(0, 0, 0)):

    palavras = texto.split()

    linhas = []
    linha_atual = ""

    for palavra in palavras:

        teste = linha_atual + " " + palavra if linha_atual else palavra

        largura = fonte_usada.render(
            teste,
            True,
            cor
        ).get_width()

        if largura <= largura_max:
            linha_atual = teste
        else:
            linhas.append(linha_atual)
            linha_atual = palavra

    if linha_atual:
        linhas.append(linha_atual)

    for i, linha in enumerate(linhas):

        desenhar_texto_centralizado(
            linha,
            y + i * 40,
            fonte_usada,
            cor
        )


def desenhar_texto_botao(texto, rect, fonte_usada=fonte, cor=(255, 255, 255)):

    img = fonte_usada.render(texto, True, cor)

    texto_rect = img.get_rect(center=rect.center)

    tela.blit(img, texto_rect)




# =========================
# DESAFIO 1 - TRAVESSIA
# =========================

esquerda = ["Raposa", "Galinha", "Alface", "Fazendeiro"]
direita = []

lado_fazendeiro = "esquerda"

movimentos = 0

mensagem_vitoria = ""
mensagem_derrota = ""


def reiniciar_travessia():
    global esquerda
    global direita
    global lado_fazendeiro
    global movimentos
    global mensagem_vitoria
    global mensagem_derrota
    global estado

    esquerda = ["Raposa", "Galinha", "Alface", "Fazendeiro"]
    direita = []

    lado_fazendeiro = "esquerda"

    movimentos = 0

    mensagem_vitoria = ""
    mensagem_derrota = ""

    estado = "travessia_menu"


def verificar_estado():
    global estado
    global mensagem_vitoria
    global mensagem_derrota

    if (
        "Raposa" in esquerda
        and "Galinha" in esquerda
        and "Fazendeiro" not in esquerda
    ):
        estado = "derrota"
        mensagem_derrota = "A raposa comeu a galinha!"
        return

    if (
        "Raposa" in direita
        and "Galinha" in direita
        and "Fazendeiro" not in direita
    ):
        estado = "derrota"
        mensagem_derrota = "A raposa comeu a galinha!"
        return

    if (
        "Galinha" in esquerda
        and "Alface" in esquerda
        and "Fazendeiro" not in esquerda
    ):
        estado = "derrota"
        mensagem_derrota = "A galinha comeu a alface!"
        return

    if (
        "Galinha" in direita
        and "Alface" in direita
        and "Fazendeiro" not in direita
    ):
        estado = "derrota"
        mensagem_derrota = "A galinha comeu a alface!"
        return

    if len(direita) == 4:
        som_vitoria.play()
        estado = "vitoria"
        mensagem_vitoria = (
            f"Parabéns! Você venceu em {movimentos} movimentos."
        )


def atravessar(item=None):
    global lado_fazendeiro
    global movimentos

    if estado != "travessia_jogando":
        return

    if lado_fazendeiro == "esquerda":

        if item:
            if item not in esquerda:
                return

            esquerda.remove(item)
            direita.append(item)

        esquerda.remove("Fazendeiro")
        direita.append("Fazendeiro")

        lado_fazendeiro = "direita"

    else:

        if item:
            if item not in direita:
                return

            direita.remove(item)
            esquerda.append(item)

        direita.remove("Fazendeiro")
        esquerda.append("Fazendeiro")

        lado_fazendeiro = "esquerda"

    movimentos += 1

    verificar_estado()
    

# =========================
# DESAFIO 2 - QUEBRA-CABEÇA
# =========================

TAMANHO_GRADE = 4
TAMANHO_PECA = 90

INICIO_X = 320
INICIO_Y = 140

grade_quebra = []
movimentos_quebra = 0

tempo_limite_quebra = 180
tempo_inicio_quebra = 0


def embaralhar_grade():
    grade = [
        [
            (i * TAMANHO_GRADE + j + 1) % (TAMANHO_GRADE * TAMANHO_GRADE)
            for j in range(TAMANHO_GRADE)
        ]
        for i in range(TAMANHO_GRADE)
    ]

    movimentos_possiveis = [(0, 1), (0, -1), (1, 0), (-1, 0)]

    vazio_l = TAMANHO_GRADE - 1
    vazio_c = TAMANHO_GRADE - 1

    for _ in range(100):
        random.shuffle(movimentos_possiveis)

        for dl, dc in movimentos_possiveis:
            novo_l = vazio_l + dl
            novo_c = vazio_c + dc

            if 0 <= novo_l < TAMANHO_GRADE and 0 <= novo_c < TAMANHO_GRADE:
                grade[vazio_l][vazio_c] = grade[novo_l][novo_c]
                grade[novo_l][novo_c] = 0

                vazio_l = novo_l
                vazio_c = novo_c
                break

    return grade


def reiniciar_quebra_cabeca():
    global grade_quebra
    global movimentos_quebra
    global estado

    grade_quebra = embaralhar_grade()
    movimentos_quebra = 0
    
    

    estado = "quebra_menu"


def desenhar_grade_quebra():
    pygame.draw.rect(
        tela,
        (220, 180, 60),
        (
            INICIO_X - 5,
            INICIO_Y - 5,
            TAMANHO_PECA * 4 + 10,
            TAMANHO_PECA * 4 + 10,
        ),
    )

    for linha in range(TAMANHO_GRADE):
        for coluna in range(TAMANHO_GRADE):
            valor = grade_quebra[linha][coluna]

            x = INICIO_X + coluna * TAMANHO_PECA
            y = INICIO_Y + linha * TAMANHO_PECA

            if valor != 0:
                pygame.draw.rect(
                    tela,
                    (35, 110, 55),
                    (x + 2, y + 2, TAMANHO_PECA - 4, TAMANHO_PECA - 4),
                )

                texto = fonte_quebra.render(str(valor), True, (255, 255, 255))
                rect_texto = texto.get_rect(
                    center=(
                        x + TAMANHO_PECA // 2,
                        y + TAMANHO_PECA // 2,
                    )
                )

                tela.blit(texto, rect_texto)


def obter_posicao_vazio():
    for l in range(TAMANHO_GRADE):
        for c in range(TAMANHO_GRADE):
            if grade_quebra[l][c] == 0:
                return l, c


def mover_peca_quebra(clique_l, clique_c):
    vazio_l, vazio_c = obter_posicao_vazio()

    if (
        abs(clique_l - vazio_l) == 1
        and clique_c == vazio_c
    ) or (
        abs(clique_c - vazio_c) == 1
        and clique_l == vazio_l
    ):
        grade_quebra[vazio_l][vazio_c], grade_quebra[clique_l][clique_c] = (
            grade_quebra[clique_l][clique_c],
            grade_quebra[vazio_l][vazio_c],
        )

        return True

    return False


def verificar_vitoria_quebra():
    valor_esperado = 1

    for l in range(TAMANHO_GRADE):
        for c in range(TAMANHO_GRADE):
            if l == TAMANHO_GRADE - 1 and c == TAMANHO_GRADE - 1:
                return grade_quebra[l][c] == 0

            if grade_quebra[l][c] != valor_esperado:
                return False

            valor_esperado += 1

    return True

# =========================
# DESAFIO 3 - ENIGMAS
# =========================

enigma_pool = [
    ("O que nasce grande e morre pequeno?", "O lapis", "O ser humano", "A arvore"),
    ("Tenho chaves, mas nao abro portas. Tenho espaco, mas nao tenho salas?", "O teclado", "O hotel", "O chaveiro"),
    ("O que corre mas nao tem pernas, e tem leito mas nao dorme?", "O rio", "O vento", "O trem"),
    ("Faco duas pessoas a partir de uma. O que eu sou?", "O espelho", "A sombra", "A foto"),
    ("Se voce me soltar eu quebro, mas se sorrir para mim eu sorrio de volta.", "O espelho", "O vidro", "A agua"),
    ("O que esta sempre a caminho, mas nunca chega de fato?", "O amanha", "O carteiro", "O futuro distante"),
    ("Quanto mais voce tira de mim, maior eu fico. O que eu sou?", "Um buraco", "A areia", "A idade"),
    ("O que tem muitas palavras, mas nunca fala nada?", "O livro", "O dicionario", "A mente"),
    ("Estou sempre no passado, criado no presente, mudo o futuro.", "A escolha", "O vento", "O pensamento"),
    ("Tenho cidades, mas nao casas. Tenho montanhas, mas nao arvores.", "O mapa", "O espaco", "O deserto"),
]

perguntas_enigma = []
pergunta_atual_enigma = None
respostas_enigma = []

indice_enigma = 0
pontuacao_enigma = 0

resposta_escolhida = None
mostrar_resultado = False

TOTAL_ENIGMAS = 5

botoes_respostas_enigma = []

botao_proxima_enigma = pygame.Rect(390, 560, 220, 50)


def reiniciar_enigma():
    global perguntas_enigma
    global indice_enigma
    global pontuacao_enigma
    global resposta_escolhida
    global mostrar_resultado
    global estado

    perguntas_enigma = list(enigma_pool)
    random.shuffle(perguntas_enigma)

    indice_enigma = 0
    pontuacao_enigma = 0

    resposta_escolhida = None
    mostrar_resultado = False

    carregar_pergunta_enigma()

    estado = "enigma_menu"


def carregar_pergunta_enigma():
    global pergunta_atual_enigma
    global respostas_enigma
    global botoes_respostas_enigma
    global resposta_escolhida
    global mostrar_resultado

    resposta_escolhida = None
    mostrar_resultado = False

    pergunta, correta, errada1, errada2 = perguntas_enigma[indice_enigma]

    respostas_enigma = [correta, errada1, errada2]
    random.shuffle(respostas_enigma)

    indice_correto = respostas_enigma.index(correta)

    pergunta_atual_enigma = (pergunta, respostas_enigma, indice_correto)

    botoes_respostas_enigma = [
    pygame.Rect(235, 360, 530, 60),
    pygame.Rect(235, 425, 530, 60),
    pygame.Rect(235, 490, 530, 60),
        ]


def verificar_resposta_enigma(indice):
    global resposta_escolhida
    global mostrar_resultado
    global pontuacao_enigma

    if mostrar_resultado:
        return

    resposta_escolhida = indice
    mostrar_resultado = True

    if indice == pergunta_atual_enigma[2]:
        pontuacao_enigma += 10


def proxima_pergunta_enigma():
    global indice_enigma
    global estado

    indice_enigma += 1

    if indice_enigma >= TOTAL_ENIGMAS:
        som_vitoria.play()
        estado = "enigma_fim"
    else:
        carregar_pergunta_enigma()



# =========================
# DESAFIO 4 - SPACE MATH
# =========================

PRETO_SPACE = (10, 10, 25)
BRANCO_SPACE = (255, 255, 255)
AZUL_SPACE = (0, 230, 255)
LARANJA_SPACE = (255, 100, 0)
AMARELO_SPACE = (255, 220, 0)
ROXO_SPACE = (90, 70, 110)

fonte_space = pygame.font.SysFont("Arial", 32, bold=True)
fonte_space_pequena = pygame.font.SysFont("Arial", 24, bold=True)

nave_x = 470
nave_y = 520
nave_vel_x = 0
tiro_space = None

pontos_space = 0
vidas_space = 3
acertos_space = 0

a_space = 0
b_space = 0
correta_space = 0
operacao_space = "+"
asteroides_space = []


def nova_pergunta_space():
    global a_space, b_space, correta_space
    global operacao_space, asteroides_space

    # Fase 1 - Soma
    if acertos_space < 10:
        operacao_space = "+"

        a_space = random.randint(1, 15)
        b_space = random.randint(1, 15)
        correta_space = a_space + b_space

    # Fase 2 - Subtração
    elif acertos_space < 20:
        operacao_space = "-"

        a_space = random.randint(10, 25)
        b_space = random.randint(1, a_space)
        correta_space = a_space - b_space

    # Fase 3 - Multiplicação
    elif acertos_space < 30:
        operacao_space = "x"

        a_space = random.randint(2, 10)
        b_space = random.randint(2, 10)
        correta_space = a_space * b_space

    # Fase 4 - Divisão
    else:
        operacao_space = "÷"

        correta_space = random.randint(2, 10)
        b_space = random.randint(2, 10)
        a_space = correta_space * b_space

    respostas = [correta_space]

    while len(respostas) < 3:
        erro = correta_space + random.randint(-5, 5)

        if erro >= 0 and erro != correta_space and erro not in respostas:
            respostas.append(erro)

    random.shuffle(respostas)

    asteroides_space = []

    posicoes_x = [180, 460, 740]
    random.shuffle(posicoes_x)

    for x, valor in zip(posicoes_x, respostas):
        asteroides_space.append({
            "rect": pygame.Rect(x, random.randint(-160, -70), 85, 85),
            "valor": valor
        })

def reiniciar_space():
    global nave_x, nave_y, tiro_space
    global pontos_space, vidas_space, acertos_space, estado

    nave_x = 470
    nave_y = 520
    nave_vel_x = 0
    tiro_space = None

    pontos_space = 0
    vidas_space = 3
    acertos_space = 0

    nova_pergunta_space()

    estado = "space_menu"


def desenhar_foguete_space(x, y):
    pygame.draw.polygon(tela, LARANJA_SPACE, [(x + 20, y + 65), (x + 40, y + 65), (x + 30, y + 90)])
    pygame.draw.ellipse(tela, BRANCO_SPACE, (x + 12, y, 36, 70))
    pygame.draw.polygon(tela, LARANJA_SPACE, [(x + 13, y + 18), (x + 47, y + 18), (x + 30, y - 5)])
    pygame.draw.circle(tela, AZUL_SPACE, (x + 30, y + 30), 8)


def desenhar_asteroide_space(asteroide):
    rect = asteroide["rect"]
    img_rect = asteroide_img.get_rect(center=rect.center)

    tela.blit(asteroide_img, img_rect)

    texto = fonte_space.render(str(asteroide["valor"]), True, AMARELO_SPACE)
    texto_rect = texto.get_rect(center=rect.center)
    tela.blit(texto, texto_rect)


def atualizar_space():
    global nave_x, tiro_space
    global pontos_space, vidas_space, acertos_space, estado

    teclas = pygame.key.get_pressed()

    # Movimento direto da nave
    if teclas[pygame.K_LEFT]:
            nave_x -= 8

    if teclas[pygame.K_RIGHT]:
            nave_x += 8

    nave_x = max(0, min(LARGURA - 60, nave_x))

    # Movimento do tiro mais controlado
    if tiro_space:
        tiro_space.y -= 6

        if tiro_space.y < 0:
            tiro_space = None

    asteroide_fugiu = False

    # Movimento dos asteroides
    for asteroide in asteroides_space:
        asteroide["rect"].y += 2

        if asteroide["rect"].y > ALTURA:
            asteroide_fugiu = True

        if tiro_space and tiro_space.colliderect(asteroide["rect"]):

            if asteroide["valor"] == correta_space:
                pontos_space += 10
                acertos_space += 1

                if pontos_space >= 400:
                    som_vitoria.play()
                    estado = "space_vitoria"
                    return
            
            else:
                vidas_space -= 1

            tiro_space = None

            if vidas_space <= 0:
                som_derrota.play()
                estado = "space_game_over"
            else:
                nova_pergunta_space()

            break

    if asteroide_fugiu and estado == "space_jogando":
        vidas_space -= 1

        if vidas_space <= 0:
            estado = "space_game_over"
        else:
            nova_pergunta_space()

# =========================
# BOTÕES
# =========================

botoes_menu = {
    "Desafio 1 - Travessia": pygame.Rect(330, 280, 340, 50),
    "Desafio 2 - Quebra-cabeça": pygame.Rect(330, 350, 340, 50),
    "Desafio 3 - Enigmas": pygame.Rect(330, 420, 340, 50),
    "Desafio 4 - Space Math": pygame.Rect(330, 490, 340, 50),
    "Sair": pygame.Rect(330, 570, 340, 50),
}

botoes_travessia = {
    "Raposa": pygame.Rect(380, 140, 220, 45),
    "Galinha": pygame.Rect(380, 200, 220, 45),
    "Alface": pygame.Rect(380, 260, 220, 45),
    "Sozinho": pygame.Rect(380, 320, 220, 45),
}

# =========================
# LOOP PRINCIPAL
# =========================

clock = pygame.time.Clock()

while True:
    clock.tick(45)

    tela.fill((173, 216, 230))

    # =========================
    # TELAS
    # =========================

    if estado == "menu_principal":

        tocar_musica(musica_menu)

        tela.blit(fundo_menu, (0, 0))
        
                
        mouse = pygame.mouse.get_pos()

        for nome, rect in botoes_menu.items():

            if rect.collidepoint(mouse):
                cor_fundo = (35, 55, 120)
                cor_borda = (255, 220, 0)
            else:
                cor_fundo = (10, 20, 60)
                cor_borda = (0, 200, 255)

            # Fundo do botão
            pygame.draw.rect(
                tela,
                cor_fundo,
                rect,
                border_radius=15
            )

            # Borda
            pygame.draw.rect(
                tela,
                cor_borda,
                rect,
                4,
                border_radius=15
            )

            # Texto
            desenhar_texto_botao(
                nome,
                rect,
                fonte,
                (255, 255, 255)
            )

    elif estado == "travessia_menu":

        tela.blit(fundo_travessia_menu, (0, 0))

        #desenhar_texto("TRAVESSIAS IMPOSSÍVEIS", 250, 40, fonte_titulo)
        #desenhar_texto("O fazendeiro precisa atravessar o rio.", 170, 140)
        #desenhar_texto("Regras:", 170, 190)
        #desenhar_texto("A raposa não pode ficar sozinha com a galinha.", 170, 250)
        #desenhar_texto("A galinha não pode ficar sozinha com a alface.", 170, 300)
        #desenhar_texto("O barco leva o fazendeiro e mais um passageiro.", 170, 350)
        #desenhar_texto("Pressione ENTER para iniciar.", 170, 470)
        #desenhar_texto("Pressione ESC para voltar ao menu principal.", 170, 520)

    elif estado == "travessia_jogando":

        tocar_musica(musica_travessia)

        tela.blit(fundo_travessia_jogando, (0, 0))

        #desenhar_texto("Margem Esquerda", 120, 40)
        #desenhar_texto("Margem Direita", 700, 40)

        posicoes_esquerda = [145, 230, 315, 400]
        posicoes_direita = [145, 230, 315, 400]

        for i, item in enumerate(esquerda):
            if i < len(posicoes_esquerda):
                desenhar_texto(item, 95, posicoes_esquerda[i])

        for i, item in enumerate(direita):
            if i < len(posicoes_direita):
                desenhar_texto(item, 780, posicoes_direita[i])

        for nome, rect in botoes_travessia.items():

            pygame.draw.rect(
                tela,
                (10, 20, 60),
                rect,
                border_radius=12
            )

            pygame.draw.rect(
                tela,
                (0, 200, 255),
                rect,
                3,
                border_radius=12
            )

            desenhar_texto(
                nome,
                rect.x + 20,
                rect.y + 10,
                cor=(255, 255, 255)
            )

        #desenhar_texto(f"Movimentos: {movimentos}", 400, 420)
        #desenhar_texto(f"Barco na: {lado_fazendeiro}", 380, 470)
        #desenhar_texto("ESC - Voltar ao menu", 380, 530)

    elif estado == "vitoria":

        tela.blit(fundo_vitoria_travessia, (0, 0))

        desenhar_texto(
            str(movimentos),
            520,
            492 ,
            fonte_titulo,
            (80, 255, 80)
        )

        #desenhar_texto(
            #mensagem_vitoria,
            #320,
            #500,
            #fonte
        #)

    elif estado == "derrota":

        tela.blit(fundo_derrota_travessia, (0, 0))

        #desenhar_texto(
            #mensagem_derrota,
            #300,
            #500,
            #fonte
        #)

    elif estado == "quebra_menu":
        #desenhar_texto("QUEBRA-CABEÇA NUMÉRICO", 240, 60, fonte_titulo)
        #desenhar_texto("Organize os números de 1 a 15.", 260, 170)
        #desenhar_texto("Clique em uma peça ao lado do espaço vazio para mover.", 180, 230)
        #desenhar_texto("Pressione ENTER para iniciar.", 300, 350)
        #desenhar_texto("Pressione ESC para voltar ao menu principal.", 230, 410)

        tela.blit(fundo_quebra_menu, (0, 0))

    elif estado == "quebra_jogando":

        tocar_musica(musica_quebra)

        tempo_decorrido = (
            pygame.time.get_ticks() - tempo_inicio_quebra
        ) // 1000

        tempo_restante = max(
            0,
            tempo_limite_quebra - tempo_decorrido
        )

        if tempo_restante <= 0:

            pygame.mixer.music.stop()

            som_derrota.play()

            estado = "quebra_derrota"

        tela.blit(fundo_quebra_jogando, (0, 0))

        desenhar_grade_quebra()

        minutos = tempo_restante // 60
        segundos = tempo_restante % 60

        desenhar_texto(
            f"{minutos:02}:{segundos:02}",
            350,
            575,
            cor=(255, 255, 255)
        )

        #desenhar_texto("QUEBRA-CABEÇA NUMÉRICO", 260, 40, fonte_titulo)
        #desenhar_texto(f"Movimentos: {movimentos_quebra}", 390, 510)
        #desenhar_texto("ESC - Voltar ao menu", 370, 560)

    elif estado == "quebra_vitoria":

        tela.blit(fundo_quebra_vitoria, (0, 0))
    
    elif estado == "quebra_derrota":

        tela.blit(fundo_quebra_derrota, (0, 0))

    elif estado == "enigma_menu":

        tela.blit(fundo_enigma_menu, (0, 0))

    elif estado == "enigma_jogando":

        tocar_musica(musica_enigmas)

        tela.blit(fundo_enigma_jogando, (0, 0))

        #desenhar_texto("DESAFIO DOS ENIGMAS", 300, 20, fonte_titulo)
        desenhar_texto(f"Pergunta {indice_enigma + 1}/{TOTAL_ENIGMAS}", 400, 130)
        desenhar_texto(f"Pontuação: {pontuacao_enigma}", 400, 170)

        pergunta = pergunta_atual_enigma[0]

        desenhar_texto_multilinha(
            pergunta,
            250,
            420
        )

        for i, rect in enumerate(botoes_respostas_enigma):
            cor = (0, 0, 0)

            if mostrar_resultado:
                if i == pergunta_atual_enigma[2]:
                    cor = (80, 200, 120)
                elif i == resposta_escolhida:
                    cor = (220, 80, 80)

            #pygame.draw.rect(tela, cor, rect)
            desenhar_texto(
                respostas_enigma[i],
                rect.x + 80,
                rect.y + 15,
                cor=cor
            )

        if mostrar_resultado:
            #pygame.draw.rect(tela, (220, 220, 220), botao_proxima_enigma)
            desenhar_texto(
                "Próxima ->>>",
                botao_proxima_enigma.x + 60,
                botao_proxima_enigma.y + 12,
            )

    elif estado == "enigma_fim":

        tela.blit(fundo_enigma_fim, (0, 0))
        
        desenhar_texto("FIM DO DESAFIO!", 340, 170, fonte_titulo)
        desenhar_texto(
            f"Sua pontuação final foi: {pontuacao_enigma} pontos.",
            315,
            280,
        )
        desenhar_texto("Pressione R para jogar novamente", 310, 380)
        desenhar_texto("Pressione ESC para voltar ao menu", 310, 430)

    elif estado == "space_menu":

        tela.blit(fundo_space_menu, (0, 0))

        
        desenhar_texto(
            "Pense rápido e destrua o asteroide correto",
            271,
            225,
            fonte,
            BRANCO_SPACE
        )

        desenhar_texto(
            "SETAS = mover nave",
            390,
            295,
            fonte,
            BRANCO_SPACE
        )

        desenhar_texto(
            "ESPACO = atirar",
            410,
            345,
            fonte,
            BRANCO_SPACE
        )

        desenhar_texto(
            "ENTER para iniciar",
            395,
            405,
            fonte,
            AMARELO_SPACE
        )

    elif estado == "space_jogando":

        tocar_musica(musica_space)

        tela.blit(fundo_space_jogando, (0, 0))

        atualizar_space()

        desenhar_texto(
            f"Pontos: {pontos_space}",
            20,
            20,
            fonte,
            AMARELO_SPACE
        )

        desenhar_texto(
            f"Vidas: {vidas_space}",
            800,
            20,
            fonte,
            LARANJA_SPACE
        )

        desenhar_texto(
            f"{a_space} {operacao_space} {b_space} = ?",
            370,
            80,
            fonte_titulo,
            BRANCO_SPACE
        )

        for asteroide in asteroides_space:
            desenhar_asteroide_space(asteroide)

        desenhar_foguete_space(
            int(nave_x),
            int(nave_y)
        )

        if tiro_space:
            pygame.draw.rect(
                tela,
                AZUL_SPACE,
                tiro_space
            )

    elif estado == "space_game_over":

        tela.blit(fundo_space_derrota, (0, 0))

        desenhar_texto(
            "GAME OVER",
            400,
            140,
            fonte_titulo,
            LARANJA_SPACE
        )

        desenhar_texto(
            f"Pontuacao Final: {pontos_space}",
            410,
            250,
            fonte,
            BRANCO_SPACE
        )

        desenhar_texto(
            "Pressione R para jogar novamente",
            340,
            390,
            fonte,
            AMARELO_SPACE
    )

    elif estado == "space_vitoria":

        tela.blit(fundo_space_vitoria, (0, 0))

        desenhar_texto_centralizado(
            "PARABÉNNSSS!",
            120,
            fonte_quebra,
            AMARELO_SPACE
        )

        desenhar_texto_centralizado(
            "Você concluiu todos os desafios matemáticos!",
            240,
            fonte,
            BRANCO_SPACE
        )

        desenhar_texto_centralizado(
            "Missao cumprida. Você venceu o Space Math!",
            320,
            fonte,
            AMARELO_SPACE
        )

        desenhar_texto_centralizado(
            f"Pontuacao Final: {pontos_space}",
            400,
            fonte,
            BRANCO_SPACE
        )

        desenhar_texto_centralizado(
            "Pressione R para jogar novamente",
            500,
            fonte,
            AMARELO_SPACE
        )

        desenhar_texto_centralizado(
            "Pressione ESC para voltar ao menu",
            560,
            fonte,
            BRANCO_SPACE
        )

    pygame.display.flip()

    # =========================
    # EVENTOS
    # =========================

    for evento in pygame.event.get():

        if evento.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

        if evento.type == pygame.KEYDOWN:

            if evento.key == pygame.K_ESCAPE:
                estado = "menu_principal"

            elif estado == "travessia_menu" and evento.key == pygame.K_RETURN:
                estado = "travessia_jogando"

            elif estado in ["vitoria", "derrota"] and evento.key == pygame.K_r:
                reiniciar_travessia()

            elif estado == "quebra_menu" and evento.key == pygame.K_RETURN:
                estado = "quebra_jogando"
                tempo_inicio_quebra = pygame.time.get_ticks()

            elif estado == "quebra_vitoria" and evento.key == pygame.K_r:
                reiniciar_quebra_cabeca()
            
            elif estado == "quebra_derrota" and evento.key == pygame.K_r:
                reiniciar_quebra_cabeca()

            elif estado == "enigma_menu" and evento.key == pygame.K_RETURN:
                estado = "enigma_jogando"

            elif estado == "enigma_fim" and evento.key == pygame.K_r:
                reiniciar_enigma()

            elif estado == "space_menu" and evento.key == pygame.K_RETURN:
                estado = "space_jogando"

            elif estado in ["space_game_over", "space_vitoria"] and evento.key == pygame.K_r:
                reiniciar_space()

            elif estado == "space_jogando" and evento.key == pygame.K_SPACE:

                if tiro_space is None:
                    tiro_space = pygame.Rect(
                        nave_x + 28,
                        nave_y - 15,
                        5,
                        22
                    )

        if evento.type == pygame.MOUSEBUTTONDOWN:

            if estado == "menu_principal":

                if botoes_menu["Desafio 1 - Travessia"].collidepoint(evento.pos):
                    som_clique.play()
                    reiniciar_travessia()

                elif botoes_menu["Desafio 2 - Quebra-cabeça"].collidepoint(evento.pos):
                    som_clique.play()
                    reiniciar_quebra_cabeca()

                elif botoes_menu["Desafio 3 - Enigmas"].collidepoint(evento.pos):
                    som_clique.play()
                    reiniciar_enigma()

                elif botoes_menu["Desafio 4 - Space Math"].collidepoint(evento.pos):
                    som_clique.play()
                    reiniciar_space()

                elif botoes_menu["Sair"].collidepoint(evento.pos):
                    som_clique.play()
                    pygame.quit()
                    sys.exit()

            elif estado == "travessia_jogando":

                if botoes_travessia["Raposa"].collidepoint(evento.pos):
                    atravessar("Raposa")

                elif botoes_travessia["Galinha"].collidepoint(evento.pos):
                    atravessar("Galinha")

                elif botoes_travessia["Alface"].collidepoint(evento.pos):
                    atravessar("Alface")

                elif botoes_travessia["Sozinho"].collidepoint(evento.pos):
                    atravessar()

            elif estado == "quebra_jogando":

                x, y = evento.pos

                if (
                    INICIO_X <= x < INICIO_X + TAMANHO_PECA * 4
                    and INICIO_Y <= y < INICIO_Y + TAMANHO_PECA * 4
                ):
                    coluna = (x - INICIO_X) // TAMANHO_PECA
                    linha = (y - INICIO_Y) // TAMANHO_PECA

                    if mover_peca_quebra(linha, coluna):
                        som_clique2.play()
                        movimentos_quebra += 1

                        if verificar_vitoria_quebra():

                            pygame.mixer.music.stop()

                            som_vitoria.play()

                            estado = "quebra_vitoria"

            elif estado == "enigma_jogando":

                for i, rect in enumerate(botoes_respostas_enigma):
                    if rect.collidepoint(evento.pos):
                        verificar_resposta_enigma(i)

                if (
                    mostrar_resultado
                    and botao_proxima_enigma.collidepoint(evento.pos)
                ):
                    proxima_pergunta_enigma()
```
